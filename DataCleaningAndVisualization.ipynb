Solution:

Histogram Plotting Attributes wise.

```
# For entire dataset, plot 4 histograms for sepal length (cm), sepal width (cm), petal length (cm) and petal width (cm)
plt.style.use('seaborn')
def plot_hist(attribute):
  
  nrows = len(attribute.columns)

  fig, axis = plt.subplots(nrows = nrows, ncols=1)
  x = 3*nrows
  y = 5*nrows

  fig.set_size_inches(x,y)

  for i in range(nrows):
    sns.histplot(attribute[attribute.columns[i]], ax = axis[i])
    axis[i].set_title(attribute.columns[i])
    plt.tight_layout();
```

```
plot_hist(data_iris_attributes)
```

Infrences:
1. Sepal length is a bimodal distribution.
2. Sepal width is somewhat seems to be a normal kind of distribution about       (3-3.25), and roughly symmetric about it.
3. Rest 2 attributes (petal length and width) are unimodal plots.

Solution:

Histogram function class_wise for each attribute.

```
def hist_plot_class(attributes):
  
  nrows = len(attributes.columns)
  
  fig, axis = plt.subplots(nrows = nrows, ncols = 1)
  fig.set_size_inches(12,22)

  for i in range(nrows):
    sns.histplot(attributes[attributes.columns[i]].loc[0:50], ax = axis[i], alpha=0.5)
    sns.histplot(attributes[attributes.columns[i]].loc[50:100], color='g', ax = axis[i],alpha=0.5)
    sns.histplot(attributes[attributes.columns[i]].loc[100:150], color = 'r', ax = axis[i],alpha=0.5)
    
    axis[i].set_title(attributes.columns[i][:-4],)
    axis[i].legend(title = attributes.columns[i][:-4], labels = [attributes.columns[i][:-4] + '_class_0',attributes.columns[i][:-4] + '_class_1', attributes.columns[i][:-4] + '_class_2'])
    plt.tight_layout()
  
```

```
# Data set contains samples 0:50 of class 0, 50:100 of class 1, 100:150 of class 2
hist_plot_class(data_iris_attributes)

# For entire dataset, plot 4 graphs for sepal length (cm), sepal width (cm), petal length (cm) and petal width (cm).
# Each graph will have three differently colored (3 colors for 3 classes) histograms.
```

**Inference:**

1. For petal length and petal width, we can clearly see the classwise distribution. Class_0 flowers basically seems to have lower petal attributes, followed by class_1 and then class_2.
2. Sepal attributes though, are somewhat mixed.

Let's study the relationship between attributes.
1. Find Pearson correlation for each pair of attributes. {Expected output is a table (triangular matrix)} **(2 points)** 
2. Also, draw a heatmap representing this correlation. **(2 points)** 

_______________________________________________________________________________________________________

Solution:

```
# Compute Pearson correlation between each pair of attributes

#creating pearson correlated table.
pearson_corr = data_iris_attributes.corr(method = 'pearson')

#transforming it to upper triangular matrix.
pearson_corr_triangular = pd.DataFrame(np.triu(pearson_corr,0), columns = pearson_corr.columns)

# The table is going to be symmetric, so you can make a triangular matrix/table 
pearson_corr_triangular
```

```
#Generate a heatmap according to the person coefficient values 
fig = plt.figure(figsize = (12,10))

#This creates a mask for printing lower triangular matrix.
mask = np.triu(np.ones_like(pearson_corr, dtype=np.bool),1)

#plotting command.
sns.heatmap(pearson_corr, annot = True, fmt = '.2f',
            cmap = 'crest',linecolor='black');

#I've plotted full heatmap. but for plotting lower triangular matrix, we just need to pass:
# 'mask=mask' inside the heatmap command above. 
```

>High(Positive) correlation means that attributes have a high positive proportional relationship. That is, if one **increases**, other **increases** too.

>Low(Negative) correlation means that attributes have a high negative or    (inverse proportional) relation. That is, if one **increases**, other **decreases**.

**Infrences:**

1. Any flower having high sepal length, has a high probability that it'll have  high petal width and petal length. Though, it doesn't depend much on sepal width. 

2. For sepal width, we don't see any high correlation value. Mostly, they are negative, which suggests, any flower with high sepal width, will have low values for other attributes.

Solution:

Function for Scatter Plotting

```
plt.style.use('seaborn')
def plot_scatter(attributes):
  
  nrows = len(attributes.columns)
  ncols = len(attributes.columns)
  fig, axis = plt.subplots(nrows = nrows, ncols=ncols)
  fig.set_size_inches(20,20)
  
  for i in range(nrows):
    for j in range(ncols):
      # if i==j:   If wan to skip diognal plots, we can comment these 2 lines.
      #   continue
      sns.scatterplot(attributes[attributes.columns[i]],attributes[attributes.columns[j]], ax = axis[i][j])
      plt.xlabel(attributes.columns[i])
      plt.ylabel(attributes.columns[j])
      plt.tight_layout()
```

```
# Generate a scatter pairplot
plot_scatter(data_iris_attributes)
```

**Inferences:**
1. Relationship between petal length and petal width, is somewhat linear. Though,there are some values missing in rand 2-3. But overall, the plot dipicts a linear relationship.

**BOX PLOT**

Box Plot is one of the most statistically representative plot. It, represents:
1. Median
2. Inter Quartile range
3. Outliers

>The blue-box represents the interquartile range (25-50) percentile.

>Points beyond the whiskers are identified as outliers.

>The middle line represents the median.

Function of Box_Plot

```
def box_plots(attributes):
  nrows = 1
  ncols = len(attributes.columns)
  fig, axis = plt.subplots(nrows = nrows, ncols=ncols)
  fig.set_size_inches(25,5)
  for i in range(ncols):
    sns.boxplot(attributes[attributes.columns[i]] , ax = axis[i])
    
```

Below, each cell plots a box_plot for indivisual class
